# The Ramsey Model

* Basically we endogenize savings from the Solow growth model
* households maximize discounted stream of lifetime utility generated by consumptions
* firms maximixe profits by choosing optimal factor inputs
* households are identical and not have market power --> inelastically supply one unit of labor on the labor market and earn a wage, w
* all households care for their children with perfect altruism -- justifying infinite time horizon, but allow for population growth (basically growth of size of household)
* households and firms are homogenous and thus we can present them as representative household and firm, respectively.

### Comparative statics

* increasing population growth will not lead to capital delusion (as in the case of Solow Model), here households anticipate faster growth of population by saving more and compensating for additional members of households

* if rate of depreciation increases --> consumption per unit and capital will both fall as in Solow model 

* if impatience increases --> save less --> capital stock per unit of effective labour will decrease and so would consumption per unit of effective labour

* change in theta (larger interest rate is needed to induce change in consumption savings per household) --> stronger desire of individuals to smooth consumption, the less they are willing to accept higher future consumption for less consumptions today --> they save less --> leads to lower steady state capital stock

## adjustment to the new steady state
 if steady state chaanges, then there is also a new saddle path on which the economy converges towards the new steady stat

## Conclusions 
- The Ramsey model generates a behavior of the economy that fits nicely to the stylized facts (convergence and long-run growth at rate A)
- In contrast to Solow, consumers are rational and optimally respond to changes in the economic environment
- One problem remain from Solow: Long-run growth is postulated rather than explained
